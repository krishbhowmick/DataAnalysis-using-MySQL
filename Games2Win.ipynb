```python
# Query #0: Find all of the records from bought table.


sql= "SELECT * FROM bought limit 10"
cursor.execute(sql)
myresult= cursor.fetchall()   #fetchall() method, which fetches all rows from the last executed statement.

for x in myresult:
  print(x)
```

```python
# Query #1.1: Total unique users who bought something, 

sql= "SELECT DISTINCT id FROM bought WHERE iap_price>0"
cursor.execute(sql)
myresult= cursor.fetchall()   #fetchall() method, which fetches all rows from the last executed statement.

for x in myresult:
  print(x)
```

```python
# Query #1.2: how many items did they buy on average 


sql= "SELECT AVG(item) FROM (SELECT id, count(item_name) AS item FROM bought GROUP BY id) SUB"
cursor.execute(sql)
myresult= cursor.fetchall()   #fetchall() method, which fetches all rows from the last executed statement.

for x in myresult:
  print(x)
```

```python
# Query #1.3: and the average price point of each item bought.



sql= "SELECT item_name, AVG(iap_price) AS avg_price FROM bought GROUP BY item_name"
cursor.execute(sql)
myresult= cursor.fetchall()   #fetchall() method, which fetches all rows from the last executed statement.

for x in myresult:
  print(x)
```

```python
# Query #2.1: Total items sold and the average price point of the items overall.



sql= "SELECT COUNT(item_name) AS total_items_sold FROM bought"
cursor.execute(sql)
myresult= cursor.fetchall()   #fetchall() method, which fetches all rows from the last executed statement.

for x in myresult:
  print(x)
```

```python
# Query #2.2: Total items sold and the average price point of the items overall.



sql= "SELECT AVG(iap_price) AS avg_price_of_overall_items FROM bought"
cursor.execute(sql)
myresult= cursor.fetchall()   #fetchall() method, which fetches all rows from the last executed statement.

for x in myresult:
  print(x)
```

```python
# Query #3: Top 3 items sold daily and revenue from these Top 3 items daily.



sql= "SELECT * FROM (SELECT ROW_NUMBER() OVER (PARTITION BY date  ORDER BY count(item_name) DESC) AS rnk, date, item_name, count(item_name) AS count_item, sum(iap_price) AS revenue FROM bought  GROUP BY date, item_name) SUB WHERE rnk<=3"
cursor.execute(sql)
myresult= cursor.fetchall()   #fetchall() method, which fetches all rows from the last executed statement.

for x in myresult:
  print(x)
```

```python
# Query #4: Total revenue made from each item broken down by country and day.



sql= "SELECT date, country, item_name, SUM(iap_price) AS revenue FROM bought GROUP BY date, country, item_name"
cursor.execute(sql)
myresult= cursor.fetchall()   #fetchall() method, which fetches all rows from the last executed statement.

for x in myresult:
  print(x)
```
