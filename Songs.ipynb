```python
# Query #1: Find all of the records of those users who have listened to Rihanna


sql= "SELECT * FROM listenings WHERE artist= 'Rihanna'"
cursor.execute(sql)
myresult= cursor.fetchall()   #fetchall() method, which fetches all rows from the last executed statement.

for x in myresult:
  print(x)
```
